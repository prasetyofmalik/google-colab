### Imports

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

### HTTP Request

#### store website in variable

```python
website = 'https://www.cars.com/shopping/results/?stock_type=cpo&makes%5B%5D=mercedes_benz&models%5B%5D=&list_price_max=&maximum_distance=20&zip='
```

#### Get Request

```python
response = requests.get(website)
```

#### Status Code

```python
response.status_code
```

### Soup Object

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
soup
```

### Results

```python
results = soup.find_all('div', {'class':'vehicle-card'})
```

```python
results[0]
```

### Put everything together inside a For-Loop

```python
# create empty lists
name = []
mileage = []
dealer_name = []
rating = []
rating_count = []
price = []

for result in results:

  # name
  try:
    name.append(result.find('h2').get_text())
  except:
    name.append('n/a')

  # mileage
  try:
    mileage.append(result.find('div',{'class':'mileage'}).get_text())
  except:
    mileage.append('n/a')

  # dealer name
  try:
    dealer_name.append(result.find('div',{'class':'dealer-name'}).get_text().strip())
  except:
    dealer_name.append('n/a')

  # rating
  try:
    rating.append(result.find('span',{'class':'sds-rating__count'}).get_text())
  except:
    rating.append('n/a')
  
  # rating count
  try:
    rating_count.append(result.find('span',{'class':'sds-rating__link'}).get_text())
  except:
    rating_count.append('n/a')
  
  # price
  try:
    price.append(result.find('span',{'class':'primary-price'}).get_text())
  except:
    price.append('n/a')
  
```

### Create Pandas DataFrame

```python
car_dealer = pd.DataFrame({'Name': name,
                           'Mileage': mileage,
                           'Dealer Name': dealer_name,
                           'Rating': rating,
                           'Rating Count': rating_count,
                           'Price': price})
```

```python
car_dealer
```

### Data Cleaning

```python
car_dealer['Rating Count'] = car_dealer['Rating Count'].apply(lambda x:x.strip('reviews)').strip('('))
```

```python
car_dealer
```

### Output in Excel

```python
car_dealer.to_excel('cardealer_single_page.xlsx', index=False)
```

### Part 2 - Pagination - Scrape 20 Pages

```python
# create empty lists
name = []
mileage = []
dealer_name = []
rating = []
rating_count = []
price = []

# website link pages
for i in range(1,21): # 21 excluded, 1-20
  # website in a variable
  website = 'https://www.cars.com/shopping/results/?page=' + str(i) + '&page_size=20&list_price_max=&makes[]=mercedes_benz&maximum_distance=20&models[]=&stock_type=cpo&zip='

  # request to website
  response = requests.get(website)

  # soup object
  soup = BeautifulSoup(response.content, 'html.parser')

  # results
  results = soup.find_all('div', {'class':'vehicle-card'})

  # iteration
  for result in results:

    # name
    try:
      name.append(result.find('h2').get_text())
    except:
      name.append('n/a')

    # mileage
    try:
      mileage.append(result.find('div',{'class':'mileage'}).get_text())
    except:
      mileage.append('n/a')

    # dealer name
    try:
      dealer_name.append(result.find('div',{'class':'dealer-name'}).get_text().strip())
    except:
      dealer_name.append('n/a')

    # rating
    try:
      rating.append(result.find('span',{'class':'sds-rating__count'}).get_text())
    except:
      rating.append('n/a')
    
    # rating count
    try:
      rating_count.append(result.find('span',{'class':'sds-rating__link'}).get_text())
    except:
      rating_count.append('n/a')
    
    # price
    try:
      price.append(result.find('span',{'class':'primary-price'}).get_text())
    except:
      price.append('n/a')
```

```python
car_dealer = pd.DataFrame({'Name': name,
                           'Mileage': mileage,
                           'Dealer Name': dealer_name,
                           'Rating': rating,
                           'Rating Count': rating_count,
                           'Price': price})
```

```python
car_dealer
```

### Data Cleaning

```python
car_dealer['Rating Count'] = car_dealer['Rating Count'].apply(lambda x:x.strip('reviews)').strip('('))
car_dealer
```

### Output in Excel

```python
car_dealer.to_excel('cardealer_multiple_pages.xlsx', index=False)
```
